# Import data and remove dupicates


```python
import pandas as pd

#read in the imported file and read the first n, or five, rows of data from csv file

df = pd.read_csv('/BX-Books.csv',error_bad_lines=False, low_memory=False, encoding='latin-1',sep=';')
df.head()
```

```python
df.info()
df.duplicated(subset='Book-Title').sum() # check for redundant entries
```

```python
df = df.drop_duplicates(subset='Book-Title')
```

```python
assert df.duplicated(subset='Book-Title').sum() < 1 # this should pass if duplicates are down to zero
```

# Creating Sample Size


```python
# randomly sample sample_size number of uniqie entries from file
sample_size = 15000
df = df.sample(n=sample_size, replace=False, random_state=490)

df = df.reset_index()
df = df.drop('index', axis=1) # 1 for columns
```

```python
df.head() # check random sample size
```

# Data Cleaning

```python
# function removes the spaces between Author names to create one word during vectorizing process later
```

```python
def clean_text(entry):
  result = str(entry).lower()
  return(result.replace(' ', ''))
```

```python
df['Book-Author'] = df['Book-Author'].apply(clean_text)
df['Book-Title'] = df['Book-Title'].str.lower()
df['Publisher'] = df['Publisher'].str.lower()
```

```python
df.head()
```

# Condensing Data

```python
# remove extra data columns
df2 = df.drop(['ISBN', 'Year-Of-Publication', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L'], axis=1)
```

```python
# combine each entries columns into one string
df2['data'] = df2[df2.columns[1:]].apply( lambda x: ' '.join(x.dropna().astype(str)), axis=1)

print(df2['data'].head())
```

```python
df2['data'].head(15000)
```

# Vectorizing data

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(df2['data'])
```

# Building Recommender System

```python
from sklearn.metrics.pairwise import cosine_similarity
# get teh cosine sililarities between each book title
similarities = cosine_similarity(vectorized)
```

```python
print(similarities)
```

```python
df = pd.DataFrame(similarities, columns=df['Book-Title'], index=df['Book-Title']).reset_index()

df.head()
```

# MAKE RECOMMENDATIONS!

```python
input_book = 'thank you, jeeves'
# make 10 recommendations based on the input book (nlargest returns the highest value in descending order from 1-0 in our case)
recommendations = pd.DataFrame(df.nlargest(11, input_book)['Book-Title'])
# make sure teh input is not recommended
recommendations = recommendations[recommendations['Book-Title']!=input_book]
print(recommendations)
```
